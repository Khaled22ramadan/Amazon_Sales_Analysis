```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read data
df = pd.read_csv('/content/amazon.csv')
```

**Data Preprocessing**

> Cleaning data for Analysis



```python
df.head()
```

```python
# check data type and nulls
# we found a problem that all columns are "object" so we need to change the numerical columns to "int64"
df.info()
```

```python
# changing the data type for price columns to float
# remove the separators between nulls
num_columns = ['discounted_price', 'actual_price']

for column in num_columns:
    # Convert the column to string type before using .str accessor
    df[column] = df[column].astype(str).str.replace('₹', '').str.replace(',', '')
    # Convert the column to float after string operations
    df[column] = pd.to_numeric(df[column], errors='coerce').astype('float')

```

```python
# clean the columns for rating and change the data type to float
df['rating'] = df['rating'].replace('|', '0').astype('float')
df['rating_count'] = df['rating_count'].fillna(0).str.replace(',', '').astype('float')
```

```python
# clean the column that contain the percentage of discount to use it in analysis
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype('float')
```

```python
# check changes
df.info()
```

```python
# Check Duplicates
duplicates = df.duplicated().sum()
duplicates
```

```python
df.head()
```

```python
df.describe()
```

```python
# check nulls
df.isnull().sum()
```

```python
# replace the nulls in rating _count with mean
mean = df['rating_count'].mean()
df['rating_count'] = df['rating_count'].fillna(mean)
```

```python
# create new_col for product name contain first 20 char from the product name column
# make it lower case in analysis
df['product_name'] = df['product_name'].str.lower()
df['product_name_short'] = df['product_name'].str[:30]
df['product_name_short'].unique()
```

```python
df['product_name']
```

```python
df.head()
```

```python
# Split category and take last part as category name for analysis
df['category'] = df['category'].str.split('|').str[-1]
df['category'].unique()
```

```python
# Category_counts
category_counts = df['category'].value_counts()
category_counts
```

```python
# Add new Features
df['price_diff'] = df['actual_price'] - df['discounted_price']  # Price difference
```

Understanding the relation between columns
